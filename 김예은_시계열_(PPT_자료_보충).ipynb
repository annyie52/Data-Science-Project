# 삼성전자 20년 vs 22년 1년치 시계열 시각화

```python
samsung_2020 = fdr.DataReader('005930', '2020-01-01', '2020-12-31').reset_index()
```

```python
samsung_2020
```

## 2020년 삼성전자 주가 흐름

```python
# 종가 기준으로 보기
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))

g = sns.lineplot(x="Date", y="Close", data=samsung_2020, palette='seismic')
g.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("종가", fontsize=15)

plt.title("2020년 일별 종가 추세", fontsize=25, pad=20)

plt.show()
```

## 2020년 삼성전자 거래량 흐름

```python
# 거래량 기준으로 보기
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))

g = sns.lineplot(x="Date", y="Volume", data=samsung_2020, palette='seismic')
g.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("거래량", fontsize=15)

plt.title("2020년 일별 거래량 추세", fontsize=25, pad=20)

plt.show()
```

```python
samsung_2020_2021["Year"] = samsung_2020_2021["Date"].dt.strftime("%Y")
samsung_2020_2021["month_day"] = samsung_2020_2021["Date"].dt.strftime("%m-%d")
```

## 2020년 2021년 한 차트에 주가 흐름 나타내기

```python
# 종가 기준으로 보기
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=45)

g = sns.lineplot(x="month_day", y="Close", hue='Year', data=samsung_2020_2021, palette='seismic')
g.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("종가", fontsize=15)

plt.title("2020년, 2021년 일별 종가 추세", fontsize=25, pad=20)

plt.show()
```

## 2020년 2021년 한 차트에 거래량 흐름 나타내기

```python
# 거래량 기준으로 보기
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=45)

g = sns.lineplot(x="month_day", y="Volume", hue='Year', data=samsung_2020_2021, palette='seismic')
g.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("종가", fontsize=15)

plt.title("2020년, 2021년 일별 거래량 추세", fontsize=25, pad=20)

plt.show()
```

## 수영복

```python
swimsuit_2018 = pd.read_excel("datalab.xlsx")
```

```python
swimsuit_2018 = swimsuit_2018.drop([0, 1, 2, 3, 4, 5], axis=0)
swimsuit_2018.columns = ["날짜", "검색량"]
swimsuit_2018 = swimsuit_2018.reset_index(drop=True)
```

### 2018년 수영복 일별 키워드 검색량 추세

```python
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=45)

g = sns.lineplot(x="날짜", y="검색량", data=swimsuit_2018, palette='seismic')
g.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("키워드 검색량", fontsize=15)

plt.title("2018년 일별 키워드 검색량 추세", fontsize=25, pad=20)

plt.show()
```

```python
swimsuit_2019
```

```python
swimsuit_2019["검색량"] = swimsuit_2019["검색량"].astype("float")
```

### 2019년 수영복 일별 키워드 검색량 추세

```python
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=45)

g = sns.lineplot(x="날짜", y="검색량", data=swimsuit_2019, palette='seismic')
g.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("키워드 검색량", fontsize=15)

plt.title("2019년 일별 키워드 검색량 추세", fontsize=25, pad=20)

plt.show()
```

```python
swimsuit_2018["Year"] = swimsuit_2018["날짜"].str[:4]
swimsuit_2018["month_day"] = swimsuit_2018["날짜"].str[5:]
```

```python
swimsuit_2019["Year"] = swimsuit_2019["날짜"].str[:4]
swimsuit_2019["month_day"] = swimsuit_2019["날짜"].str[5:]
```

```python
sns.set_style("whitegrid")
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=45)

g = sns.lineplot(x="month_day", y="검색량", data=swimsuit_2018, palette='seismic', color="blue")
k = sns.lineplot(x="month_day", y="검색량", data=swimsuit_2019, palette='seismic', color="red")
g.xaxis.set_major_locator(ticker.MultipleLocator(31))
k.xaxis.set_major_locator(ticker.MultipleLocator(31))


plt.xlabel("날짜", fontsize=15)
plt.ylabel("키워드 검색량", fontsize=15)

plt.legend(["2018년", "2019년"])

plt.title("2018년, 2019년 일별 수영복 검색량 추세", fontsize=25, pad=20)

plt.show()
```
